```python
import pandas as pd
import pickle
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
tdf = pd.read_csv('npf_train.csv')
tdf = tdf.drop(columns=['date', 'id', 'partlybad'])
tdf['class2'] = (tdf['class4'] == 'nonevent').astype(int)
xdf = tdf.drop(columns=['class4', 'class2'])
ydf = tdf['class2']
```

```python
rf = RandomForestClassifier(n_estimators=1000, random_state=42, class_weight={1: 0.55, 0:0.45}, min_samples_leaf=5, min_samples_split=5)
scaler = StandardScaler()
```

```python
xcols = xdf.columns
xdf = scaler.fit_transform(xdf)
rf.fit(X=xdf, y=ydf)
```

```python
feature_imp = pd.DataFrame({"features":xcols, "RF_importance": rf.feature_importances_})
feature_imp.sort_values('RF_importance', ascending=False).head(50)
```
