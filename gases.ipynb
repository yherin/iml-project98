```python
import pickle
import pandas as pd
from pandas import Series
import project_utils as pu
```

```python
tdf = pd.read_csv('npf_train.csv')
tdf = tdf.drop(columns=['date', 'id', 'partlybad'])
```

```python
gases = ['CO2', 'NO', 'NOx', 'O3', 'H2O', 'SO2', 'CO', 'RHIRGA', 'T' ]
units = {'CO2': 'ppm', 'NO':'ppb', 'NOx': 'ppb', 'O3': 'ppb', 'H2O': 'ppt', 'SO2': 'ppb', 'CO': 'ppb', 'RHIRGA': 'ppm', 'T': 'ppm'}

gasdict = {}

def gasregex_mean(g):
    return  f'{g}\d+\.mean'

def gasregex_std(g):
    return  f'{g}\d+\.std'

def convert_ppm(g):
    unit =  units[g.name.split('_')[0]]
    print(g.name)
    print(g)
    print(unit)
    if unit == 'ppm':
        return g
    elif unit == 'ppt':
        return g * 1000
    elif unit == 'ppb':
        return g / 1000
    else:
        raise ValueError(f'invalid unit {unit}')

```

```python
for g in gases:
    gasdict[f'{g}_mean'] = tdf.filter(regex=gasregex_mean(g), axis=1).mean(axis=1)
    gasdict[f'{g}_std'] = tdf.filter(regex=gasregex_std(g), axis=1).mean(axis=1)
```

```python
gmdf = pd.DataFrame(gasdict)
gmdf.head()
```

```python
gmdf = gmdf.apply(convert_ppm, axis=0)
gmdf.head()
```

```python
pickle.dump(gmdf, open('gases_agg_dataframe.bin', 'wb'))
```
